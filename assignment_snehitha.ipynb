3. Using the date column, create new columns for: year, month, day of the week and hour of the day.

```python
df['year'] = df['datetime'].dt.year
df['month'] = df['datetime'].dt.month
df['day_of_week'] = df['datetime'].dt.dayofweek
df['hour'] = df['datetime'].dt.hour
df[['year','month','day_of_week','hour']].head()

```

11. Create a new categorical column called day_period, which can take four possible values: night, morning, afternoon and evening. These values correspond to the following binning of the hour column: 0-6: night, 6-12: morning, 12-6: afternoon, 6-24:evening.


```python
def get_day_period(hour):
    if 0 <= hour < 6:
        return 'night'
    elif 6 <= hour < 12:
        return 'morning'
    elif 12 <= hour < 18:
        return 'afternoon'
    else:
        return 'evening'

df['day_period'] = df['hour'].apply(get_day_period)
df[['hour','day_period']].head()
```

12. Generate a pivot table for the mean of the hourly total rentals count, with the index set to the day period and the column set to the working day column.
What can you observe from the table ?

```python
pivot_table = pd.pivot_table(
    df,
    values='count',
    index='day_period',
    columns='workingday',
    aggfunc='mean'
)

pivot_table
```

Observations from this pivot:

Bike rentals are highest in the morning and evening on working days (commute times), while on non-working days afternoon rentals are higher. Night rentals are low for both working and non-working days.
